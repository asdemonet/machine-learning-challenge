```python
# Import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("starter_code/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_disposition', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_slogg', 'koi_srad']]
selected_features
```

```python
# Set X and y
X = selected_features[['koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_slogg', 'koi_srad']]
y = selected_features['koi_disposition']
print(X.shape, y.shape)
```

```python
# Model Type: Random Forest
from sklearn.preprocessing import LabelEncoder
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original: ' + str(original_class))
    print('Encoded: ' + str(label))
    print('----------------')
```

```python
from tensorflow.keras.utils import to_categorical
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
# Create a train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=42, test_size=0.3)
```

```python
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
scaled_X_train = X_scaler.transform(X_train)
scaled_X_test = X_scaler.transform(X_test)
X_scaler
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=600)
model.fit(scaled_X_train, y_train)
model.score(scaled_X_test, y_test)
```

```python
print(f"Training Data Score: {model.score(scaled_X_train, y_train)}")
print(f"Testing Data Score: {model.score(scaled_X_test, y_test)}")
```

```python
sorted(zip(model.feature_importances_, df.columns), reverse=True)
```
